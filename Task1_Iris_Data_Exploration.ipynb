# Task 1: Exploring and Visualizing Iris Dataset

## Problem Statement
The objective of this task is to explore and analyze the Iris dataset using data visualization techniques.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

iris = sns.load_dataset("iris")
iris.head()
```

```python
# Shape of dataset
print("Shape of dataset:", iris.shape)

# Column names
print("\nColumns:", iris.columns)

# Dataset information
iris.info()

# Statistical summary
iris.describe()
```

```python
# Check for missing values
iris.isnull().sum()
```

```python
plt.figure()
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
iris.hist()
plt.show()
```

```python
plt.figure()
sns.boxplot(data=iris)
plt.title("Boxplot of Iris Features")
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

## Final Insights

- The dataset contains 150 samples and 4 numerical features.
- There are no missing values.
- Petal length and petal width clearly separate species.
- Setosa is easily distinguishable from others.
- The dataset is clean and well-balanced.
